## Machine Learning section 

Deep learning models have been used since there are able to read complex patherns in the data and therefore make more accurate predictions that standard machine learning models. Thus, a ANN model will be performed on this dataset and compared with a traditional machine learning classifier to measure both performances.

## **Data Preparation**

**Select X and y**
First time this model was tested, all variable were used, but it resulted in an overfitting model. That was the reason 
that variables were reduced as shown in next step

```python
# Select X and y
# This was used to test all the features in the first time this experiment was performed. 
#The model was overfitted using all variables
#X = Credit.loc[:, Credit.columns != 'Class']
#y = Credit.loc[:, 'Class']
```

```python
# Select X and y 
# 17 variables were selected instead
X = Credit.loc[:, [ 'V6', 'V7', 'V8', 'V9','V10', 'V11', 'V12','V13', 'V14', 'V15','V16', 'V17', 'V18', 'V19', 'V20','V21','V22']] # independent variables

y = Credit['Class']# dependent variable
```

```python
X
```

```python
y
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 1, stratify=y) #58 #test_size = 0.25, random_state = 0
```

```python
# Performing preprocessing part
from sklearn.preprocessing import StandardScaler

# Create and initialise an object (sc) by calling a method named as StandardScaler()
sc = StandardScaler()

# Train the model by calling a method fit_transform()
X_train = sc.fit_transform(X_train)

# Transform the data into standised form
X_test = sc.transform(X_test)
```

```python
#y
```

## Builiding an Artificial Neural Network model (ANN)

```python
from keras.layers import Dense
from keras.models import Sequential

# Build neural network in Keras
model = Sequential()

# Add fully connected layer with a ReLU activation function
model.add(Dense(10, input_dim=17, activation='relu', )) # 10

# Add fully connected layer with a ReLU activation function
model.add(Dense(2, activation='relu')) #2 units

# Add fully connected layer with a sigmoid activation function - Output layer
model.add(Dense(1, activation='sigmoid')) # output layer
```

```python
#Checking summary of the model
model.summary()
```

```python
#Compile neural network
# loss = binary_crossentropy for binary classification problems
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy']) 
```

```python
#Fitting the model with 25 epochs, batch size 100
model.fit(X_train, y_train, epochs=25, batch_size=100)#, verbose=False)  loss: 0.0426 - accuracy: 0.9849
```

```python
# Results - Accuracy
scores = model.evaluate(X_train, y_train) # verbose = False)
print("Training Accuracy: %.2f%%\n" % (scores[1] * 100))  # last time model was tested loss: 0.0412 - accuracy: 0.9855

scores = model.evaluate(X_test, y_test) # verbose = False) loss 
print("Testing Accuracy: %.2f%%\n" % (scores[1] * 100)) #  loss: 0.0415 - accuracy: 0.9856
```

```python
# Evaluating the predictions of the model
from sklearn.metrics import confusion_matrix
import seaborn as sns

y_test_pred = model.predict(X_test)
y_test_pred = y_test_pred.flatten()
y_test_pred_new = np.where(y_test_pred.round(2) > 0.5, 1, 0)
c_matrix = confusion_matrix(y_test, y_test_pred_new)
sns.heatmap(c_matrix, annot=True)
c_matrix
```

**Discussion:** Last time the model was tested: I had 113,726 number of transactions in my test set. Out of 55,693 were predicted to be legitimate transactions, and 1224 were classified as being legitimate, when  there were not (type one error). It correctly predicted 55,639 transactions as being product of fraud and 1,170 as being not legitimate ones, when they really are (type two error).

```python
#Checking the perfomance of the model with a confussion matrix
import seaborn as sns
# Predicting the Test set results
y_pred = model.predict(X_test).flatten().round()

#Verify the accuracy of the model
from sklearn.metrics import accuracy_score
print('ANN accuracy:',accuracy_score(y_test,y_pred))

# Classification report 
from sklearn.metrics import f1_score , precision_score , recall_score , classification_report
#print('classification_report:',classification_report(y_test,y_pred))
print('f1_score:',f1_score(y_test,y_pred))
print('precision_score:',precision_score(y_test,y_pred))
print('recall_score:',recall_score(y_test,y_pred))
```

**Discussion:** As seen above, the model loss has been reduced when using 25 epochs and acuracy remains between 98% producing better results. 

```python
# Get training and test loss histories
history = model.fit(X_train, y_train, epochs=25, verbose=0, validation_data=(X_test, y_test))
```

```python
#Visualizing the loss in train set 
#plt.figure(figsize=(6,5))
plt.title('Loss') 
plt.plot(history.history['loss'], label='train') 
plt.plot(history.history['val_loss'], label='test') 
plt.legend(["Training Loss", "Test Loss"])
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show();
```

**Alternatively, we can use the same approach to visualize the training and test accuracy over each epoch**

```python
# plot accuracy during training 

plt.title('Accuracy') 
plt.plot(history.history['accuracy'], label='train') 
plt.plot(history.history['val_accuracy'], label='test') 
plt.legend(["Training Accuracy", "Test Accuracy"])
plt.xlabel("Epoch")
plt.ylabel("Accuracy Score")
plt.show() 
```

## Compare ANN with Random Forest Classifier

A Random Forest classifier is going to be trained to compared the results with ANN model. Train test and test set for X and y are going to be the same.

```python
#Import Random Forest Model
from sklearn.ensemble import RandomForestClassifier

#Create a Gaussian Classifier
clf = RandomForestClassifier(n_estimators = 50)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred)) ##0.9795
```

```python
#Visualizing f1_score , precision_score , recall_score values
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Calculate cm by calling a method named as 'confusion_matrix'
cm = confusion_matrix(y_test, y_pred)

# Call a method heatmap() to plot confusion matrix
sns.heatmap(cm, annot = True)

# print the classification_report based on y_test and y_predict
print(classification_report(y_test, y_pred))

# Classification report 
from sklearn.metrics import f1_score , precision_score , recall_score , classification_report
print('f1_score:',f1_score(y_test,y_pred))
print('precision_score:',precision_score(y_test,y_pred))
print('recall_score:',recall_score(y_test,y_pred))
```

**Discussion:** After performing ANN and RF classifier, was possible to see that ANN produced more accurate predictions while Random forest model was overfitted while tested with the same parameters for training and testing sets. As demostrated on previous research, when using big data, traditional machine learning models do not perform well. In contrast, Deep learning model, as ANN in this case produced better results to identify the number of fraudulent transactions.
